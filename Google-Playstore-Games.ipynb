### <span style="color:#89CFF0">4 - Analiz Öncesi Gerekli Düzenlemeler</span>

```python
df_1 = DataFrame.copy()
```

```python
df_1.head(1)
```

#### Amaç: installs sütunundaki dtype'ı object olan verileri float değerlere dönüştürerek data analizinde işimizi kolaylaştırmak.

```python
df_1.installs.value_counts()
```

```python
def in_thousand (inst) :
    if inst == '500.0 k' :
        return '0.5 M'
    elif inst == '100.0 k' :
        return '0.1 M'
    else :
        return inst
```

```python
df_1['installs'] = df_1['installs'].apply(in_thousand)
```

```python
df_1['installs']
```

```python
df_1['installs'] = df_1['installs'].str.replace( 'M', '').str.strip().astype('float')
```

```python
df_1['installs']
```

```python
df_1 = df_1.rename(columns = {'installs': 'installs_in_million'})
```

```python
df_1['installs_in_million'].value_counts().sort_index()
```

```python
DataFrame.head(1)
```

#### Amaç: Oyunların yıldız alma sayıları elimizde ancak daha doğru bir sonuca ulaşmak için yıldızları yüzde olarakta veri setimizde görmek istiyoruz.

```python
df_1['5 star percent'] = round(df_1['5 star ratings']  / (df_1['5 star ratings'] + df_1['4 star ratings'] + df_1['3 star ratings'] + df_1['2 star ratings'] + df_1['1 star ratings']) * 100, 2)
df_1['4 star percent'] = round(df_1['4 star ratings']  / (df_1['5 star ratings'] + df_1['4 star ratings'] + df_1['3 star ratings'] + df_1['2 star ratings'] + df_1['1 star ratings']) * 100, 2)
df_1['3 star percent'] = round(df_1['3 star ratings']  / (df_1['5 star ratings'] + df_1['4 star ratings'] + df_1['3 star ratings'] + df_1['2 star ratings'] + df_1['1 star ratings']) * 100, 2)
df_1['2 star percent'] = round(df_1['2 star ratings']  / (df_1['5 star ratings'] + df_1['4 star ratings'] + df_1['3 star ratings'] + df_1['2 star ratings'] + df_1['1 star ratings']) * 100, 2)
df_1['1 star percent'] = round(df_1['1 star ratings']  / (df_1['5 star ratings'] + df_1['4 star ratings'] + df_1['3 star ratings'] + df_1['2 star ratings'] + df_1['1 star ratings']) * 100, 2)
```

### <span style="color:#89CFF0">6 - Sonuç</span>

#### <span style="color:#89CFF0">Özet</span>
##### Mobil oyunlar hakkında bazı istatistikler incelendiğinde, çoğunluğu 10 milyon indirme alıyor. En çok indirilen iki oyun türü Arcade ve Casual olarak belirlendi. En beğenilen oyun türü Educational iken en beğenilmeyen oyun türü Sport oldu. Oyunların çoğunluğu ücretsizken, ücretli oyunlar genelde Arcade ve Music türünde yapılmıştır. Ücretli oyunların ücreti ortalama 2.7 civarındadır ve en yüksek ücret 7.5, en düşük ücret ise 1'dir. Ücretsiz oyunların indirilme sayısı, ücretli oyunların indirilme sayısına göre daha fazladır. Ücretli oyunlar ortalama 2.01 milyon kez indirilirken, ücretsiz oyunlar ortalama 29.29 milyon kez indiriliyor.

##### Son 30 günde en çok gelişen oyun türü Casino iken, arkasından Trivia ve Card geliyor. Son 60 günde ise en çok gelişen oyun türü Card, arkasından Board ve Strateji geliyor. En çok indirilen oyunların türü Casual ve Arcade iken, en fazla indirilen 100 oyun arasından kategorisinde en çok oyun barındıran tür Arcade olarak belirlendi. Ortalama puanı 4'ün altında olan oyunlar 100 milyon indirme barajını geçememiştir.

##### En çok 5 yıldız yüzdesine sahip oyun türü Casino iken, en çok 1 yıldız yüzdesine sahip oyun türü Music oyunlarıdır. Anormal yükseliş gösteren oyun türleri Casino iken, anormal düşüş gösteren oyun türü Board'dur.

#### <span style="color:#89CFF0">Öneriler</span>

##### 1-) Arcade veya Casual türü oyunlar yapabilirsiniz çünkü en çok indirilen oyun türleri bunlar arasından seçilmiş.

##### 2-) Educational türü oyunlar da beğenilme açısından üst sıralarda yer aldığı için, bu türde bir oyun yapabilirsiniz.

##### 3-) Ücretli oyunların genellikle Arcade ve Music türünde yapıldığından bahsedilmiş, bu türlerde ücretli oyunlar yapabilirsiniz.

##### 4-) Ortalama puanı 4'ün altında olan oyunların 100 milyon indirme barajını geçemediği belirtilmiş, bu nedenle kaliteli bir oyun yapmak önemlidir.

##### 5-) Son 30 günde en çok gelişen oyun türü Casino iken, son 60 günde en çok gelişen oyun türü Card olmuş. Bu trendlere göre, Casino veya Card türünde oyunlar yapmayı düşünebilirsiniz.

##### 6-) Anormal yükseliş gösteren oyun türleri Casino iken, anormal düşüş gösteren oyun türü Board olarak belirlenmiş. Bu nedenle, Board türünde oyunlar yapmaktan kaçınabilirsiniz.

##### 7-) Son olarak, mümkünse ücretsiz bir oyun yapmayı düşünebilirsiniz çünkü ücretsiz oyunlar genellikle daha fazla indiriliyor ve oyununuzun daha geniş bir kitleye ulaşmasını sağlayabilirsiniz.
